## Compare Tissue Samples in DLPFC Dataset

```python
from hest import iter_hest
import matplotlib.pyplot as plt

ids = [f"MISC{i}" for i in range(1,13)]
ids 
```

```python
samples = iter_hest(hest_dir="../hest_data",id_list=ids)
```

### Plot tissue Images

```python
fig, axes = plt.subplots(3,4, figsize = (10,10))
for i,sample in enumerate(samples):
    x,y = i % 4,i // 4
    ax = axes[y,x]
    ax.imshow(sample.adata.uns['spatial']['ST']['images']['downscaled_fullres'])
    ax.set_title(f"{sample.meta['dataset_title']} sample {sample.meta['subseries']} \nHest ID: {sample.meta['id']}")
fig.tight_layout()
```
